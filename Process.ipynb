Model

```python
#Design Model
class Model():
    def __init__(self,pix_json_object , *args, **kargs):
        super().__init__(*args, **kargs)
        root_path = '/content/pix3d.json'

        self.pix_obj = pix_json_object
        self.mesh = trimesh.load(root_path + self.pix_obj['model'],force='mesh')
        self.keypoitns_2d = self.pix_obj['2d_keypoints']

        with open(root_path + self.pix_obj['3d_keypoints'], 'r') as file:
            self.keypoints_3d = file.read().split('\n')[:-2]

        self.mask = np.asarray(Image.open(root_path + self.pix_obj['mask']))
        self.img = np.asarray(Image.open(root_path + self.pix_obj['img']))

    def keyPoints2d(self):
        x=[]
        y=[]
        for i in self.keypoitns_2d:
            for j in i:
                x.append(j[0])
                y.append(j[1])
        return (x, y)


    def keyPoints3d(self):
        x=[]
        y=[]
        z=[]
        for i in self.keypoints_3d:
            i = i.split(' ')
            x.append(float(i[0]))
            y.append(float(i[1]))
            z.append(float(i[2]))

        return (x, y, z)


    def showMesh(self):
        return self.mesh.show()

    def vertices(self):
        vertices = np.asarray(self.mesh.vertices)
        x=[]
        y=[]
        z=[]

        for i in vertices:
            x.append(i[0])
            y.append(i[1])
            z.append(i[2])

        return (x ,y , z)

    def showImg(super, put_mask = False):
        x, y = super.keyPoints2d()
        if put_mask:
            plt.scatter(x,y,c='blue')
            plt.imshow(super.mask)
        else:
            plt.scatter(x,y,c='blue')
            plt.imshow(super.img)

        return None

    def showPlanes(super,plotModel=False):

        ax1 = plt.figure(figsize=(3,3)).add_subplot()
        ax2 = plt.figure(figsize=(3,3)).add_subplot()
        ax3 = plt.figure(figsize=(3,3)).add_subplot()

        x_3d,y_3d,z_3d = super.keyPoints3d()
        x, y, z = super.vertices()

        if plotModel:
            ax1.set_title('xy plane') # xy plane
            ax1.set_xlabel('x')
            ax1.set_ylabel('y')
            ax1.plot(x,y,c='red')
            ax1.scatter(x_3d,y_3d,c='blue')

            ax2.set_title('yz plane') # zy plane
            ax2.set_xlabel('z')
            ax2.set_ylabel('y')
            ax2.plot(z,y,c='red')
            ax2.scatter(z_3d,y_3d,c='blue')

            ax3.set_title('xz plane') # zx plane
            ax3.set_xlabel('z')
            ax3.set_ylabel('x')
            ax3.plot(z,x,c='red')
            ax3.scatter(z_3d,x_3d,c='blue')

        else:
            ax1.set_title('xy plane') # xy plane
            ax1.set_xlabel('x')
            ax1.set_ylabel('y')
            ax1.scatter(x_3d,y_3d,c='blue')

            ax2.set_title('yz plane') # zy plane
            ax2.set_xlabel('z')
            ax2.set_ylabel('y')
            ax2.scatter(z_3d,y_3d,c='blue')

            ax3.set_title('xz plane') # zx plane
            ax3.set_xlabel('z')
            ax3.set_ylabel('x')
            ax3.scatter(z_3d,x_3d,c='blue')

        return None

    def showIn3d(super):

        x_3d,y_3d,z_3d = super.keyPoints3d()
        x, y, z = super.vertices()

        ax = plt.figure().add_subplot(projection='3d')

        ax.plot(x,y,z,c='blue')
        ax.scatter(x_3d,y_3d,z_3d,c='red')
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.set_zlabel('z')

        return None


```
